```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
le=LabelEncoder()
data['gender']=lb.fit_transform(data['gender'])
data=pd.get_dummies(data,columns=['occasion','receiver_personality','relationship_to_receiver','interests','giver_preference_style'])
budget_map={'below_500':0,'500_to_1000':1,'1000_to_3000':2,'3000_to_5000':3}
data['budget']=data['budget'].map(budget_map)
data['gift_type']=le.fit_transform(data['gift_type'])
data['gift_name']=le.fit_transform(data['gift_name'])
receiver_age_group_map={'Child':0,'Teen':1,'Adult':2,'Senior':3}
data['receiver_age_group']=data['receiver_age_group'].map(receiver_age_group_map)
relationship_closeness_map={'distant':0,'neutral':1,'close':2,'very close':3}
data['relationship_closeness']=data['relationship_closeness'].map(relationship_closeness_map)
```
